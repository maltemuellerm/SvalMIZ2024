```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy.spatial.distance import pdist, squareform
from sklearn.neighbors import DistanceMetric

# Assuming StatisticalEvaluation class and related functions are already defined elsewhere
buoyno = [23]
startdate = datetime.strptime('2024-04-04 00:00:00', '%Y-%m-%d %H:%M:%S')
enddate = datetime.strptime('2024-05-22 00:00:00', '%Y-%m-%d %H:%M:%S')
ifiletemp = '/home/maltem/work/python/data/SvalMIZ2024/colocatedFiles/updated_dataset_temp.nc'
ifiledrift = '/home/maltem/work/python/data/SvalMIZ2024/colocatedFiles/dataset_trajectories_full.nc'

# Use StatisticalEvaluation to get the dataset
s1 = StatisticalEvaluation(buoyno, 'ECMWF-AIFS', startdate, enddate, [1, 1], ifiledrift, ifiletemp)
s1.info()
s1.flag_values(rad_threshold=120)

bmax = 32
tmax = int(np.shape(s1.OMBtemp.time_ds.values)[0])
temp_series_data = np.zeros([bmax, tmax])

# Collecting the temperature, lat, and lon data
for buoyno in range(32):
    temp_1m = np.squeeze(s1.OMBtemp.temp_1m_calibrated[buoyno, :].values)
    lat_buoy = np.squeeze(s1.OMBtemp.lat_obs[buoyno, :].values)
    lon_buoy = np.squeeze(s1.OMBtemp.lon_obs[buoyno, :].values)
    flag = np.squeeze(s1.OMBtemp.temp_flag[buoyno, :].values.astype(bool))

    # Assign temperature data to the matrix, applying the flag
    temp_series_data[buoyno, :] = temp_1m - 273.15  # Convert from Kelvin to Celsius
    temp_series_data[buoyno, flag] = np.nan

# Count the number of valid time series (non-flagged) at each time point
valid_series_count = np.sum(~np.isnan(temp_series_data), axis=0)

# Calculate the ensemble mean and standard deviation at each time point, ignoring NaNs
ensemble_mean = np.nanmean(temp_series_data, axis=0)
ensemble_std = np.nanstd(temp_series_data, axis=0)

# Create a time vector (for example, just using the indices of the points)
time = s1.OMBtemp.time_ds.values

# Calculate the mean distance between the buoys at each time point
# Get the lat/lon data for all buoys
coords = np.column_stack((lat_buoy, lon_buoy))
distance_matrix = squareform(pdist(coords, metric='euclidean'))  # Calculate distances in lat-lon space

# Compute the mean distance for each time point (just an example; adjust based on data)
mean_distance = np.nanmean(distance_matrix, axis=0)
std_distance = np.nanstd(distance_matrix, axis=0)

# Set up the plots
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 12), sharex=True,
                                          gridspec_kw={'height_ratios': [3, 1, 1, 1]})

# Plot the ensemble mean and spread
ax1.plot(time, ensemble_mean, color='k', label='Ensemble Mean', linewidth=2)
ax1.fill_between(time, ensemble_mean - ensemble_std, ensemble_mean + ensemble_std,
                 color='blue', alpha=0.3, label='1 Std Dev Range')
ax1.fill_between(time, ensemble_mean - 2 * ensemble_std, ensemble_mean + 2 * ensemble_std,
                 color='blue', alpha=0.1, label='2 Std Dev Range')
ax1.set_ylabel('Temperature (°C)')
ax1.set_title('Ensemble Mean with Spread (Std Dev)')
ax1.legend()

# Plot the number of valid (non-flagged) time series
ax2.plot(time, valid_series_count, color='orange', label='Valid Time Series', linewidth=2)
ax2.set_xlabel('Time')
ax2.set_ylabel('Number of Time Series')
ax2.set_title('Valid Time Series Count')
ax2.legend()

# Plot the mean distance between buoys over time
ax3.plot(time, mean_distance, color='green', label='Mean Distance', linewidth=2)
ax3.fill_between(time, mean_distance - std_distance, mean_distance + std_distance,
                 color='green', alpha=0.3, label='Std Dev Range')
ax3.set_xlabel('Time')
ax3.set_ylabel('Mean Distance (m)')
ax3.set_title('Mean Distance Between Buoys')
ax3.legend()

# Plot the variogram (example: just based on the distance matrix and the standard deviation)
# The variogram is typically the semi-variance of the data vs. lag distance.
variogram = np.mean(distance_matrix ** 2, axis=0) / 2  # This is just a simplified calculation
ax4.plot(time, variogram, color='purple', label='Variogram', linewidth=2)
ax4.set_xlabel('Time')
ax4.set_ylabel('Semivariance')
ax4.set_title('Variogram')
ax4.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
